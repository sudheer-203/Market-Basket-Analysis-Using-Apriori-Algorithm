```python
# Import relevant panda
import pandas as pd
```

```python
# Read the data from UCI
data = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
```

```python
# Print the information from the data
data.info()
```

```python
# Checking the original data
data.head()
```

```python
# Include relevant data by removing incomplete observations
data = data.dropna()
# Remove Invoice with Credit Values
data['InvoiceNo'] = data['InvoiceNo'].astype('str')
data = data[~data['InvoiceNo'].str.contains('C')]

# Re-check the data information after removal 
data.info()
```

```python
# Checking the countries 
data['Country'].unique()
```

```python
# Checking how many observations for each country
data['Country'].value_counts()
```

## UK

```python
# Subset the observations in UK
transaction_data_UK = (data[data['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
# Checking the UK data
transaction_data_UK.head()
```

```python
# Make a function to convert the unit as 'x' to 1 hot encoded values, 1 when the product is exist in the transaction
def hotencode(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1    
```

```python
# Apply the function into the data
transaction_data_UK_encoded = transaction_data_UK.applymap(hotencode)
```

```python
transaction_data_UK_encoded
```

```python
# To find the relationship between product we need the data of transaction with more than one item
finaltransaction_UK = transaction_data_UK_encoded[(transaction_data_UK_encoded > 0).sum(axis = 1) >= 2]
```

```python
finaltransaction_UK.head()
```

```python
# Build up the frequent items
from mlxtend.frequent_patterns import apriori
frequent_itemsets_UK = apriori(finaltransaction_UK, min_support=0.02, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
```

```python
frequent_itemsets_UK['length'] = frequent_itemsets_UK['itemsets'].apply(lambda x: len(x))
frequent_itemsets_UK
```

```python
# Create the rules
from mlxtend.frequent_patterns import association_rules
rules_UK = association_rules(frequent_itemsets_UK, metric='lift', 
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```

```python
rules_UK[ (rules_UK['confidence'] >= 0.7) & (rules_UK['support'] >= 0.02)]
```

```python
finaltransaction_UK['PINK REGENCY TEACUP AND SAUCER'].sum()
```
